## Pregunta 2 

```python
## cual seria box plot de  median PESO de sexo 1
```

```python
x = pd.pivot_table(df,
            index = ['desc_ent'],
            columns = ['sexo'],
            values = ['peso'],
            aggfunc = {'peso':['median']}
              )
x.head()
```

```python
x.columns
```

```python
sns.boxplot(data = x[('peso', 'median', 1)])
```
