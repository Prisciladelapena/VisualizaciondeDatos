```python
df= pd.read_excel('datasets/Canada.xlsx',
                  sheet_name='Canada by Citizenship',
                  engine='openpyxl',
                  skiprows=range(20),
                  skipfooter=200
                 )
df              
```

```python
df.columns=list(map(str,df.columns))
all(isinstance(column, str) for column in df.columns)
```

```python
all(isinstance(column, str) for column in df.columns)
```

```python
df.columns
```

```python
df.drop(['Type' ,'Coverage', 'REG', 'DEV', 'DevName', 'Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46',
       'Unnamed: 47', 'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'], inplace=True,
     axis=1)
df.columns
     
```

```python
df.rename(columns={'OdName': 'Country',
                   'AreaName': 'Continent',
                   'RegName': 'Region',
                  },
          inplace=True)
df.columns
```

```python
df['Total']=df.iloc[:,5:].sum(axis=1)
df.head()
df.set_index('Country', inplace=True)
df.head()
```

Pregunta de investigacion: #7
Me gustaria saber cuantas personas llegaron desde Haiti a Canada desde 1980'-2013, con una grafica de lineas


```python
df.loc['Haiti', years].plot()
plt.xlabel('Años', fontsize=16)
plt. xticks(fontsize=16, rotation=45)

plt.ylabel('Numero de imigrantes', fontsize=16)
plt.yticks(fontsize=16, rotation=45)
plt.title('Numero de imigrantes desde Haiti a Canada desde 1980-2013', fontsize=17)
plt.text(30,6400,'Terremoto')
plt.show()
```
