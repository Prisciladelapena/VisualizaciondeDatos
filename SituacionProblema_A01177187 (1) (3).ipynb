```python
#modelo de prediccion entre ensenanza e investigacion
y=ensenanza
x=investigacion

#Agrega una constante = 1 que es el multiplicador del intercepto (True - al inicio)
constant = sm.tools.tools.add_constant(x,prepend=True,has_constant='skip')

result = sm.OLS(y,constant).fit()

print(result.summary())
```

```python
#Para generar predicciones
result.predict()
```

```python
sm.graphics.plot_fit(result,1, vlines=False);
```

```python
#modelo de prediccion entre ensenanza y alumnos internacionales
y2=ensenanza
x2=intestudiantes

#Agrega una constante = 1 que es el multiplicador del intercepto (True - al inicio)
constant2 = sm.tools.tools.add_constant(x2,prepend=True,has_constant='skip')

result = sm.OLS(y2,constant2).fit()

print(result.summary())
```

```python
result.predict()
```

```python
#predicciones del modelo 
predict = result.predict()

residuals = ensenanza - predict

print(residuals)
```

```python
#grafica de dispersion de residuales
import matplotlib.pyplot as plt

plt.scatter(ensenanza,residuals)

plt.show()
```
